```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('spam.csv')
df.head()
```

```python
df.Category.value_counts()
```

```python
df['Spam'] = df['Category'].apply(lambda x: 1 if x=='spam' else 0)
df.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test, y_train, y_test = train_test_split(df['Message'], df.Spam, test_size=0.2)
```

```python
X_train.shape
```

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer()
X_train_cv = cv.fit_transform(X_train)
```

```python
# Model building (Naive Bayes classifier)
```

```python
from sklearn.naive_bayes import MultinomialNB

nb = MultinomialNB()
nb.fit(X_train_cv, y_train)
```

```python
X_test_cv = cv.transform(X_test)
```

```python
y_pred = nb.predict(X_test_cv)
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_pred,y_test))
```

```python
emails =['Hello, how are you doing? Its been long since we caught up last time. Hope you are doing good',
        'Hurrah!! you won $ 287655434 prize and its waiting in your nostro. grab the bonanza']
cn = cv.transform(emails)
nb.predict(cn)
```

```python
# alternative method

from sklearn.pipeline import Pipeline

clpip = Pipeline([
    ('vectorizer' , CountVectorizer()),
    ('nb', MultinomialNB())
])
```

```python
clpip.fit(X_train, y_train)
```

```python
print(classification_report(y_test, clpip.predict(X_test)))
```
